```python
df_select = df_decode[selected_cols].copy()
df_select.head(5)
```

```python
# Copy dataframe untuk memproses fitur baru
df_select_fe = df_select.copy()
```

```python
df_select_fe['avg_sem_approved'] = (
    df_select_fe['Curricular_units_1st_sem_approved'] + df_select_fe['Curricular_units_2nd_sem_approved']
) / 2

df_select_fe['avg_sem_grade'] = (
    df_select_fe['Curricular_units_1st_sem_grade'] + df_select_fe['Curricular_units_2nd_sem_grade']
) / 2

df_select_fe['avg_sem_without_evaluation'] = (
    df_select_fe['Curricular_units_1st_sem_without_evaluations'] + df_select_fe['Curricular_units_2nd_sem_without_evaluations']
) / 2
```

```python
df_select_fe.drop([
    'Curricular_units_1st_sem_approved',
    'Curricular_units_2nd_sem_approved',
    'Curricular_units_1st_sem_grade',
    'Curricular_units_2nd_sem_grade',
    'Curricular_units_1st_sem_without_evaluations',
    'Curricular_units_2nd_sem_without_evaluations'
], axis=1, inplace=True)
df_select_fe.head(5)
```

```python
from sklearn.model_selection import train_test_split

X = df_select_fe  # fitur numerik + kategorikal
y = df_decode['Status']     # target

y = y.map({'Dropout': 0, 'Graduate': 1})
y

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
```

```python
num_cols_model = [
    'Age_at_enrollment',
    'avg_sem_approved',
    'avg_sem_grade',
    'avg_sem_without_evaluation',
    'Application_mode'
]
```

```python
cat_cols_model = [
    'Scholarship_holder',
    'Debtor',
    'Gender',
    'Tuition_fees_up_to_date'
]
```

## Modeling

```python
!pip install imblearn
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
```

```python
# ColumnTransformer
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), num_cols_model),
    ('cat', OneHotEncoder(drop='first'), cat_cols_model)
])
```

```python
# Pipeline modeling
model_pipeline = Pipeline(steps=[
    ('preprocessing', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('classifier', LogisticRegression())
])
```

```python
# Fit model
model_pipeline.fit(X_train, y_train)
```

```python
# Prediksi
y_pred = model_pipeline.predict(X_test)
```

## Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score, classification_report

# y_test adalah label asli
# y_pred adalah hasil prediksi model
# y_proba adalah probabilitas prediksi (untuk ROC AUC)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# confusion matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# heat map cm
sns.heatmap(cm, annot=True, fmt='g', cmap="YlGnBu",xticklabels=['Dropout', 'Graduate'],
            yticklabels=['Dropout', 'Graduate'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python


# Pipeline khusus visualisasi
visual_preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), num_cols_model),
    ('cat', OneHotEncoder(drop=None), cat_cols_model)  # ⬅️ drop=None agar semua kategori terlihat
])

visual_pipe = Pipeline(steps=[
    ('preprocess', visual_preprocessor),
    ('model', LogisticRegression(max_iter=1000))
])
```

```python
visual_pipe.fit(X_train, y_train)
```

```python
# Ambil nama fitur numerik dan kategorikal
num_features = num_cols_model
cat_encoder = visual_pipe.named_steps['preprocess'].named_transformers_['cat']
cat_features = cat_encoder.get_feature_names_out(cat_cols_model)

# Gabungkan semua nama fitur
all_features = list(num_features) + list(cat_features)
```

```python
import pandas as pd
import numpy as np

coefficients = visual_pipe.named_steps['model'].coef_[0]
coef_df = pd.DataFrame({
    'Feature': all_features,
    'Coefficient': coefficients
}).sort_values(by='Coefficient', ascending=False)
coef_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(data=coef_df, x='Coefficient', y='Feature', palette='coolwarm')
plt.title('Koefisien Logistic Regression (Visualisasi Lengkap)')
plt.tight_layout()
plt.show()
```

Karena ini adalah klasifikasi binary (0 = Dropout, 1 = Graduate), maka:
•	Koefisien positif → menaikkan peluang mahasiswa lulus (Graduate)
•	Koefisien negatif → menurunkan peluang mahasiswa lulus, artinya cenderung ke Dropout


Koefisien Positif (peluang lulus meningkat):  


1. avg_sem_approved	1.40,	Semakin banyak unit semester yang disetujui (lulus), makin tinggi peluang mahasiswa Graduate

3. Scholarship_holder_Has Scholarship	0.47,	Mahasiswa yang menerima beasiswa cenderung lulus.  
4. avg_sem_grade	0.41,	Nilai rata-rata yang lebih tinggimeningkatkan peluang lulu.  
5. Debtor_No Debt	0.40,	Mahasiswa yang tidak memiliki hutang lebih berpotensi lulus.  
6. Gender_Female	0.09,	lebih besar kemungkinan mahasiswa perempuan yang lulus.





Koefisien Negatif (peluang dropout meningkat):  


1.   avg_sem_without_evaluation	-0.14,	Semakin banyak unit yang tidak dievaluasi peluang Dropout meningkat.  

2.   Age_at_enrollment	-0.20,	Semakin tua usia saat masuk kuliah maka peluang Dropout meningkat
